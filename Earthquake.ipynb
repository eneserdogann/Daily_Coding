```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

fig, axes = plt.subplots(1, 3, figsize=(16,4))

df = pd.read_csv('data.csv',parse_dates=['Date'])
df = df[df['Magnitude']>3]
series = pd.to_datetime(df['Date'], format='%Y-%m-%d %H:%M:%S')

df['Date'] = series.dt.date
year = series.dt.year
month = series.dt.month
week = series.dt.isocalendar().week

fig.subplots_adjust(hspace=0.6)

axes[0].bar(year,df['Magnitude'])
axes[0].set_xlabel("Years")
axes[0].set_ylabel("Magnitude")
axes[1].bar(month,df['Magnitude'])
axes[1].set_xlabel("Months")
axes[1].set_ylabel("Magnitude")
axes[2].bar(week,df['Magnitude'])
axes[2].set_xlabel("Weeks")
axes[2].set_ylabel("Magnitude")
```
